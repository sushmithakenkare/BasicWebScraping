```python
#Sushmitha Srinath Kenkare

from urllib.request import urlopen as req
from bs4 import BeautifulSoup as soup
import pandas as pd
import re
import csv
```

```python
#Fetching required HTML code from the Web. 

URL = 'https://h1bdata.info/index.php?em=&job=Data+Scientist&city=&year=2020' #URL where we need data from
uClient = req(URL) #Requesting Data. 
html_code = uClient.read() #Storing the retrieved HTML code.   
uClient.close() #Close the Client after the fetching is done. 
```

```python
#Parsing through the HTML page and retrieving data that we need

html_soup = soup(html_code,"html.parser") #Using BeautifulSoup for Parsing

rows = html_soup.findAll("tr") #After inspecting the HTML code, we know that the data needed is enclosed within 'tr' tags

header = rows[0] #First row contains headers
uClient.close() #Close the Client after the fetching is done. 
headings = []

for item in header.findAll("th"):
	item = (item.text).rstrip("\n") #Only getting the text part enclosed within the 'th' tag.
	headings.append(item)

all_rows = []
body = rows[1:] #Once we have headings(column names) sorted out. We now need to retrieve data to populate those columns.


for row_num in range(len(body)):
	row = []
	for row_item in body[row_num].findAll("td"): #Data that we need is enclosed within 'td' tags. 
		clean = re.sub("(\xa0)|(\n)|,","", row_item.text) #Getting rid of commas present in base salary and new-line characters(\n)
		row.append(clean)

	all_rows.append(row)

df = pd.DataFrame(data=all_rows,columns = headings) #Storing the data retrieved as a dataframe. 
df.head()
```

```python
#Saving the DataFrame to a CSV file
df.to_csv('h1b_info.csv', index=False) 
```
